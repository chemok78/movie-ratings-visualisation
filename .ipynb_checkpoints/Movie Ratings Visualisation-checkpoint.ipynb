```python
import pandas as pd
# import fivethirthyeight movies data
reviews = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/fandango/fandango_score_comparison.csv")
# select only the columns we need
columns=["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
norm_reviews=reviews[columns]
print(norm_reviews.columns)
print(norm_reviews.head(10))
```

```python
import matplotlib.pyplot as plt
from numpy import arange
# select values of first row and all columns
bar_heights = norm_reviews.ix[0,columns].values
# Positions of the left sides of the bars. [0.75, 1.75, 2.75, 3.75, 4.75]
# create a list evenly spaced between 5 elements + 0.75 for each
bar_positions = arange(5) + 1
# create a subplot and assign Figure object and Axes object
fig,ax = plt.subplots()
# create bar plot
# syntax: ax.bar(left, height, width)
# left is x coordinates of the bars
ax.bar(bar_positions, bar_heights, 0.5)
# formatting
tick_positions = range(1,6)
ax.set_xticks(tick_positions)
ax.set_xticklabels(columns, rotation=90)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

```python
# same bar chart, but horizontal
bar_widths = norm_reviews.ix[0, columns].values
bar_positions = arange(5) + 1
tick_positions = range(1,6)
fig,ax = plt.subplots()
ax.barh(bar_positions, bar_widths, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(columns)
ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title("Average User Rating for Avengers: Age of Ultron (2015)")
plt.show()
```

```python
# Scatterplot comparing Fandango to Rotten Tomatoes ratings
fig,ax = plt.subplots()
ax.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")
plt.show()
```

There is a weak positive 
correlation between Fandango and Rotten Tomatoes ratings

The correlation is weak because for many x values, there are multiple corresponding y valies. The correlation is positive because in general as x increaes, y increases as well

```python
# flip Fandango and RT columns and compare
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax2.set_xlabel("Rotten Tomatoes")
ax2.set_ylabel("Fandango")
ax2.scatter(norm_reviews["RT_user_norm"], norm_reviews["Fandango_Ratingvalue"])
plt.show()
```

```python
# comparing correlation between Fandango and RT, Metacritic and IMDB
# set x and y labels in every chart to be equal ranging from 0 to 5
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)
ax2.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["Metacritic_user_nom"])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)
ax3.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["IMDB_norm"])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB")
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)
plt.show()
```

User ratings from Fandango and IMDB are the most similar. User ratings from Metacritic and Rotten Tomatoes have positive but weak correlations with user ratings from Fandango. 

```python
# first look value distrbution Fandango and IMDB
fandango_distribution = norm_reviews["Fandango_Ratingvalue"].value_counts().sort_index()
imdb_distribution = norm_reviews["IMDB_norm"].value_counts().sort_index()
print(fandango_distribution)
print(imdb_distribution)
```

```python
# histogram for distribution of values Fandango ratings
fig,ax = plt.subplots()
ax.hist(norm_reviews["Fandango_Ratingvalue"], range=[0,5])
plt.show()
```

```python
# histrograms for Fandango, RT, Metacritics and IMDB to compare distribution
# 20 bins, 0-5 for x and 0-50 for y for all charts
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)
ax1.hist(norm_reviews["Fandango_Ratingvalue"], 20,range=(0,5))
ax1.set_title("Distribution of Fandango Ratings")
ax1.set_ylabel("Frequency")
ax2.hist(norm_reviews["RT_user_norm"], 20,range=(0,5))
ax2.set_title("Distribution of Rotten Tomatoes Ratings")
ax2.set_ylabel("Frequency")
ax3.hist(norm_reviews["Metacritic_user_nom"], 20,range=(0,5))
ax3.set_title("Distribution of Metacritic Ratings")
ax3.set_ylabel("Frequency")
ax4.hist(norm_reviews["IMDB_norm"], 20,range=(0,5))
ax4.set_title("Distribution of IMDB Ratings")
ax4.set_ylabel("Frequency")
ax1.set_ylim(0,50)
ax2.set_ylim(0,50)
ax3.set_ylim(0,50)
ax4.set_ylim(0,50)
plt.show()
```

* Around 50% of user ratings from Fandango fall in the 2 to 4 score range
* Around 50% of user ratings from Rotten Tomatoes fall in the 2 to 4 score range
* Around 75% of the user ratings from Metacritic fall in the 2 to 4 score range
* Around 90% of the user ratings from IMDB fall in the 2 to 4 score range

```python
# box plot to see the quartiles of Rotten Tomatoes
# each quartile covers an equal number of values (histogram proportion of the actual values)
fig,ax = plt.subplots()
ax.boxplot(norm_reviews["RT_user_norm"])
ax.set_ylim(0,5)
ax.set_xticklabels(["Rotten Tomatoes"])
plt.show()
```

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
fig, ax = plt.subplots()
ax.boxplot(norm_reviews[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0,5)
plt.show()
```
