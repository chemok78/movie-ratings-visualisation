```python
import pandas as pd
# import fivethirthyeight movies data
reviews = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/fandango/fandango_score_comparison.csv")
# select only the columns we need
columns=["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
norm_reviews=reviews[columns]
print(norm_reviews.columns)
print(norm_reviews.head(10))
```

```python
import matplotlib.pyplot as plt
from numpy import arange
# select values of first row and all columns
bar_heights = norm_reviews.ix[0,columns].values
# Positions of the left sides of the bars. [0.75, 1.75, 2.75, 3.75, 4.75]
# create a list evenly spaced between 5 elements + 0.75 for each
bar_positions = arange(5) + 1
# create a subplot and assign Figure object and Axes object
fig,ax = plt.subplots()
# create bar plot
# syntax: ax.bar(left, height, width)
# left is x coordinates of the bars
ax.bar(bar_positions, bar_heights, 0.5)
# formatting
tick_positions = range(1,6)
ax.set_xticks(tick_positions)
ax.set_xticklabels(columns, rotation=90)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

```python
# same bar chart, but horizontal
bar_widths = norm_reviews.ix[0, columns].values
bar_positions = arange(5) + 1
tick_positions = range(1,6)
fig,ax = plt.subplots()
ax.barh(bar_positions, bar_widths, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(columns)
ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title("Average User Rating for Avengers: Age of Ultron (2015)")
plt.show()
```
